```python
import tifffile as tiff, numpy as np, glob, json

paths = glob.glob("../DSIFN-CD/test/labels/*.tif")
vals = {}
nodata_flags = []
for p in paths[:50]:
    im = tiff.imread(p)
    tags = tiff.TiffFile(p).pages[0].tags
    nodata = None
    for tag in tags.values():
        if tag.name.lower() in ["gdal_nodata", "nodata"]:
            nodata = tag.value
    unique, counts = np.unique(im, return_counts=True)
    vals[p] = dict(zip(map(int, unique), map(int, counts)))
    nodata_flags.append(nodata)
print("nodata samples:", [n for n in nodata_flags if n is not None][:5])
print(json.dumps(vals, indent=2)[:1500])
```

```python
# view_and_convert_tif_labels.py
import numpy as np
from pathlib import Path
import tifffile as tiff
import matplotlib.pyplot as plt
from PIL import Image

LABELS_DIR = Path("../DSIFN/test/mask")   # carpeta con .tif
OUT_DIR    = Path("../DSIFN-CD/label_png_preview")
OUT_DIR.mkdir(parents=True, exist_ok=True)

def to_uint8_mask(arr):
    """
    Convierte una m√°scara cualquiera a binaria 0/255:
    - si el rango ya es 0/1 => *255
    - si es 0/255 => deja igual
    - si tiene otros valores => umbral en 128 (o mediana si prefieres)
    """
    arr = np.asarray(arr)
    arr = arr.squeeze()         # (H,W,1) -> (H,W)
    arr = arr.astype(np.float32)

    vmin, vmax = arr.min(), arr.max()
    print(f"   rango: [{vmin}, {vmax}]")

    if vmax <= 1.0:             # 0/1
        mask = (arr > 0.5).astype(np.uint8) * 255
    elif vmax <= 255.0:         # 0..255
        # normalizamos a 0/255 por seguridad (por si hay grises)
        mask = (arr >= 128).astype(np.uint8) * 255
    else:
        # valores grandes (e.g. 0..65535), umbral relativo
        thr = (vmax + vmin) / 2.0
        mask = (arr >= thr).astype(np.uint8) * 255

    return mask

def main():
    tifs = sorted([p for p in LABELS_DIR.glob("*.tif")])
    if not tifs:
        print(f"No encontr√© .tif en {LABELS_DIR}")
        return

    for i, tif_path in enumerate(tifs):  # muestra/convierte primeros 10 a modo demo
        print(f"\n[{i+1}/{len(tifs)}] {tif_path.name}")
        arr = tiff.imread(tif_path)  # soporta multi-p√°gina; usualmente 1 banda
        print("   shape:", arr.shape, "dtype:", arr.dtype)

        # Visualizar
        plt.figure(figsize=(5,5))
        if arr.ndim == 2:
            plt.imshow(arr, cmap="gray")
        else:
            # si por accidente tiene m√°s canales, muestra el primero
            plt.imshow(arr[..., 0], cmap="gray")
        plt.title(tif_path.name)
        plt.axis("off")
        plt.tight_layout()
        plt.show()

        # Guardar como PNG binario (opcional)
        mask = to_uint8_mask(arr)
        out_png = OUT_DIR / (tif_path.stem + ".png")
        Image.fromarray(mask).save(out_png)
        print(f"   PNG binario guardado en: {out_png}")

if __name__ == "__main__":
    main()
```

```python
from PIL import Image
import numpy as np
import os

def binarizar_carpeta(SPLIT, carpeta_entrada: str, carpeta_salida: str = None, extensiones=(".png", ".jpg", ".jpeg")):
    """
    Binariza todas las im√°genes en una carpeta (0/255) y guarda los resultados.
    - carpeta_entrada: ruta de las im√°genes originales.
    - carpeta_salida: carpeta donde guardar las im√°genes binarizadas (si no existe, se crea).
    - extensiones: tipos de archivo a procesar (por defecto PNG/JPG/JPEG).
    """
    if carpeta_salida is None:
        carpeta_salida = f"DSIFN-CD/{SPLIT}/label/"
    os.makedirs(carpeta_salida, exist_ok=True)

    archivos = [f for f in os.listdir(carpeta_entrada) if f.lower().endswith(extensiones)]

    if not archivos:
        print("‚ö†Ô∏è No se encontraron im√°genes en la carpeta.")
        return

    for nombre in archivos:
        ruta_in = os.path.join(carpeta_entrada, nombre)
        ruta_out = os.path.join(carpeta_salida, os.path.splitext(nombre)[0] + ".png")

        img = Image.open(ruta_in).convert("L")
        arr = np.array(img)

        min_val, max_val = arr.min(), arr.max()
        if max_val == min_val:
            bin_arr = np.zeros_like(arr, dtype=np.uint8)
        else:
            umbral = (min_val + max_val) / 2
            bin_arr = (arr > umbral).astype(np.uint8) * 255

        Image.fromarray(bin_arr, mode="L").save(ruta_out)
        print(f"‚úÖ Guardado: {ruta_out}")

    print(f"\nüéâ Binarizaci√≥n completa. Archivos guardados en: {carpeta_salida}")

# Ejemplo de uso:
binarizar_carpeta("train","DSIFN-CD/train/mask")
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

from pathlib import Path
from PIL import Image

# === CONFIG ===
SPLIT="test" # train, val, test
ROOT = Path(f"DSIFN-CD/{SPLIT}")   # carpeta ra√≠z del dataset
DELETE_OLD = True                 # cambia a True si quieres borrar los .jpg despu√©s

def convert_dir(subdir: str):
    path = ROOT / subdir
    if not path.exists():
        print(f"‚ö†Ô∏è Carpeta no encontrada: {path}")
        return
    count = 0
    for jpg_path in path.glob("*.jpg"):
        png_path = jpg_path.with_suffix(".png")
        with Image.open(jpg_path) as im:
            im.save(png_path)
        count += 1
        if DELETE_OLD:
            jpg_path.unlink()
    print(f"Convertidos {count} archivos en {subdir}/")

def main():
    for sub in ["A", "B"]:
        convert_dir(sub)
    print("üéØ Conversi√≥n completada para A y B en DSIFN-CD")

if __name__ == "__main__":
    main()
```

```python
import os
import glob

def borrar_jpg_en_carpeta(ruta_carpeta: str):
    """
    Elimina todos los archivos .jpg (y .jpeg) de la carpeta especificada.
    """
    # Busca archivos con extensi√≥n .jpg y .jpeg (insensible a may√∫sculas)
    patrones = [os.path.join(ruta_carpeta, '*.jpg'),
                os.path.join(ruta_carpeta, '*.jpeg'),
                os.path.join(ruta_carpeta, '*.JPG'),
                os.path.join(ruta_carpeta, '*.JPEG')]

    archivos = []
    for patron in patrones:
        archivos.extend(glob.glob(patron))

    # Elimina los archivos encontrados
    for archivo in archivos:
        try:
            os.remove(archivo)
            print(f"‚úÖ Borrado: {archivo}")
        except Exception as e:
            print(f"‚ö†Ô∏è No se pudo borrar {archivo}: {e}")

    print(f"\nTotal de archivos borrados: {len(archivos)}")

# Ejemplo de uso:
SPLIT = "train"
SUB = ["A","B"]

def main():
    for sub in ["A", "B"]:
        borrar_jpg_en_carpeta(f"DSIFN-CD/{SPLIT}/{sub}")
    print("jpg borrados")

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

from pathlib import Path
import re

# === CONFIG ===
DATASET_ROOT = Path("DSIFN-CD256")  # ajusta si tu ruta base es distinta
#SPLITS = ["train", "val", "test"]
SPLITS = ["train"]
def extract_num(filename: str):
    """Extrae el n√∫mero del nombre (ej. 'train_12.png' -> 12)."""
    match = re.search(r"(\d+)", filename)
    return int(match.group(1)) if match else -1

def main():
    for split in SPLITS:
        split_dir = DATASET_ROOT / split / "A"
        list_dir = DATASET_ROOT / split / "list"
        list_dir.mkdir(parents=True, exist_ok=True)

        # Archivos PNG del directorio A
        img_names = sorted(
            [p.name for p in split_dir.glob("*.png")],
            key=extract_num
        )

        list_path = list_dir / f"{split}.txt"
        with open(list_path, "w") as f:
            for name in img_names:
                f.write(name + "\n")

        print(f"[‚úì] {split}: {len(img_names)} archivos guardados en {list_path}")

if __name__ == "__main__":
    main()
```
